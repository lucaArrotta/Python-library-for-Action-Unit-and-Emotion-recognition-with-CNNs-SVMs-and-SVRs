# Testing
In this notebook we'll test the lib functions implemented in "au_lib.py" file.
1. __get_img_aus_occurencies__: Returns the list of the AUs detected in the input image
2. __get_img_emotions_occurencies__: Returns a dict in which at each emotion is associated the percentage of the emotion AUs which appear in the input image (based on occurencies)
3. __get_img_aus_intensities__: Returns a dict in which at each detected AU is associated the predicted intensity
